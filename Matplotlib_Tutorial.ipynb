### Load Necessary Libraries

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

### Basic Graph

```python
x = [0,1,2,3,4]
y = [0,2,4,6,8]

# Resize your Graph (figsize uses inches; here-8 in by 5 in) (dpi specifies pixels per inch. Using 100 here... When saving probably should use 300 if possible)
# Hence, total pixels in this graph would be 8*100 by 5*100 pixels 
plt.figure(figsize=(8,5), dpi=100)


# *** Line 1 ***

# Keyword Argument Notation
#plt.plot(x,y, label='2x', color='red', linewidth=2, marker='.', linestyle='--', markersize=10, markeredgecolor='blue')

# Shorthand notation, b^-- = <color=b> <marker=^> <linestyle=-->
# fmt = '[color][marker][line]'
plt.plot(x,y, 'b^--', label='2x') # or plt.plot(x,y, label='2x', color='red', linewidth=2, marker='.', linestyle='--', markersize=10, markeredgecolor='blue')
# check out https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.plot.html#matplotlib.pyplot.plot for the list of usable shorthand notation "MARKERS"


## *** Line 2 ***

# select interval we want to plot points at
x2 = np.arange(0,4.5,0.5) #gives a range of points
print (x2)

# Plot one part of the graph as line
# 6 is exclusive so it finishes at 5 and then we pick up at 5 in next one
plt.plot(x2[:6], x2[:6]**2, 'r', label='X^2')

# Plot remainder part of graph as a dot
plt.plot(x2[5:], x2[5:]**2, 'r--')

# Add a title (specify font parameters with fontdict)
plt.title('Our First Graph!', fontdict={'fontname': 'Comic Sans MS', 'fontsize': 20})

# X and Y labels
plt.xlabel('X Axis', fontdict={'fontname': 'Times New Roman', 'fontsize': 15})
plt.ylabel('Y Axis', fontdict={'fontname': 'Times New Roman', 'fontsize': 15})

# X, Y axis Tickmarks (scale of your graph)
plt.xticks([0,1,2,3,4,])
#plt.yticks([0,2,4,6,8,10])

# Add a legend - this displays the labels
plt.legend()

# Save figure (dpi of 300 is good when saving so graph has high resolution)
plt.savefig('mygraph.png', dpi=300)

# Show plot
plt.show()
```

### Bar Chart

```python
labels = ['A', 'B', 'C']
values = [1,4,2]

plt.figure(figsize=(5,3), dpi=100) #resizing

bars = plt.bar(labels, values) #(x axis, y axis)

patterns = ['/', 'O', '*'] #add different patterns to bars
for bar in bars:
    bar.set_hatch(patterns.pop(0)) #or, pop() to give the patterns in reverse

# or,
# bars[0].set_hatch('/')
# bars[1].set_hatch('O')
# bars[2].set_hatch('*')

plt.savefig('barchart.png', dpi=300)

# plt.legend() #not really necessary

plt.show() #removes the annoying line like <BarContainer object of 3 artists>
```

####Line Plots

```python
#Line plot over a function

fig = plt.figure(figsize=(10, 10), dpi=50)
ax2 = plt.axes(projection = '3d')

#(starting point, end poiont, step size)... i.e. start from 0 and with an increase of 0.1 each time, reach 50
x_data = np.arange(0, 50, 0.1) 
y_data = np.arange(0, 50, 0.1)
# z_data = x_data * y_data
z_data = np.sin(x_data) * np.cos(y_data)

# ax2.scatter(x_data, y_data, z_data, marker="v", alpha=0.9) #scatter functoin will plot each point from x, y, z_data
ax2.plot(x_data, y_data, z_data) #plot function will plot a line 

#Setting Labels
ax2.set_title("Random Trigo Function")
ax2.set_xlabel("X-Current (A)")
ax2.set_ylabel("Y-Resistance (Ohm)")
ax2.set_zlabel("Z-Results Voltage (A*Ohm = V)")


plt.show()
```

#### Surface Plots
How to plot surfaces, add colour maps, and how to change/rotate the default viewing angle 

```python
# Surface plot by using mesh grid
# Basically we need to multiply each x value with all y values available (rather than just multiplying x_i and y_i)
# we have used 2-D arrays to store multiple rows of the same x and y values

fig = plt.figure(figsize=(10, 10), dpi=50)
ax3 = plt.axes(projection = '3d')

#(starting point, end point, step size)... i.e. start from 0 and with an increase of 0.1 each time, reach 50
# x_data = np.arange(0, 50, 0.1) 
# y_data = np.arange(0, 50, 0.1)
# X, Y = np.meshgrid(x_data, y_data) #X, Y are our 2-D arrays ... # can print(X) to check what X stores
# Z = X * Y

x_data = np.arange(-5, 5, 0.1) #reducing ranges for a trigo function
y_data = np.arange(-5, 5, 0.1)
X, Y = np.meshgrid(x_data, y_data) #X, Y are our 2-D arrays
Z = np.sin(X) * np.cos(Y)
 
ax3.plot_surface(X, Y, Z, cmap='plasma') #plot_surface function will plot a surface
#ADDING A COLOUR MAP: Changes like a heatmap, with hieght; can choose any cmap

#ROTATING THE DEFAULT VIEWING ANGLE - using azimuth and elevation (specified in degrees)
ax3.view_init(azim = 0, elev = 90)

#Setting Labels
ax3.set_title("Random Surface Plot")
ax3.set_xlabel("X")
ax3.set_ylabel("Y")
ax3.set_zlabel("Z")


plt.show()
```

### Pie Chart
Lets draw a piechart on the percentage of people who play with their left foot vs their right foot

```python
left = fifa.loc[fifa['Preferred Foot'] == 'Left'].count()[0]
right = fifa.loc[fifa['Preferred Foot'] == 'Right'].count()[0]

plt.figure(figsize=(8,5))

labels = ['Left', 'Right']
colors = ['#abcdef', '#aabbcc']

#it gets messed up with we don't pass it numbers like plt.pie(fifa[Preferred Foot'])
plt.pie([left, right], labels = labels, colors=colors, autopct='%.3f %%') #.3f tells 3 decimal spaces; %% tells to print % character

plt.title('Foot Preference of FIFA Players')

plt.show()
```

### Pie Chart #2

Lets draw a piechart on the weights of people

```python
plt.figure(figsize=(8,5), dpi=100)

plt.style.use('ggplot') #instead of mentioning every colour of everry label ourselves, we can use some style from list of available styles

fifa.Weight = [int(x.strip('lbs')) if type(x)==str else x for x in fifa.Weight] #because weights in csv file are like 355lbs in string format; int(...) typecasts the string like '355' to int value 355
fifa.Weight #print

# WE can even convert pounds to Kilograms at this point

light = fifa.loc[fifa.Weight < 125].count()[0]
light_medium = fifa[(fifa.Weight >= 125) & (fifa.Weight < 150)].count()[0]
medium = fifa[(fifa.Weight >= 150) & (fifa.Weight < 175)].count()[0]
medium_heavy = fifa[(fifa.Weight >= 175) & (fifa.Weight < 200)].count()[0]
heavy = fifa[fifa.Weight >= 200].count()[0]

weights = [light,light_medium, medium, medium_heavy, heavy]
label = ['under 125', '125-150', '150-175', '175-200', 'over 200']

explode = (.4,.2,0,0,.4) #start off by trying... explode = (.1,.1,.1,.1,.1)

plt.title('Weight Distribution of Professional Soccer Players (lbs)')

plt.pie(weights, labels=label, explode=explode, pctdistance=0.8,autopct='%.2f %%') #pctdistance = distance from centre of graph; 0 at centre, 1 at circumference, >1 beyond circumference
plt.show()
```

### Box and Whiskers Chart

We use these to try and compare teams
1. We can see the best player up top (top whisker)
2. and worst player down bottom (bottom whisker)
3. alongwith the core of the team (25%ile to 75%ile) in the boxed part and the average

```python
plt.figure(figsize=(5,8), dpi=100)

plt.style.use('default') #need to change back the style to default after having used it in a previous cell

barcelona = fifa.loc[fifa.Club == "FC Barcelona"]['Overall']
madrid = fifa.loc[fifa.Club == "Real Madrid"]['Overall']
revs = fifa.loc[fifa.Club == "New England Revolution"]['Overall']

#bp = plt.boxplot([barcelona, madrid, revs], labels=['a','b','c'], boxprops=dict(facecolor='red'))
bp = plt.boxplot([barcelona, madrid, revs], labels=['FC Barcelona','Real Madrid','NE Revolution'], patch_artist=True, medianprops={'linewidth': 2}) #it doesn't recognise "facecolor" property by default so we use patch_artist=True
#medianprops is a property to change property of the median lne like (here) width

plt.title('Professional Football Team Comparison')
plt.ylabel('FIFA Overall Rating')

for box in bp['boxes']:
    # change outline color
    box.set(color='#4286f4', linewidth=2)
    # change fill color
    box.set(facecolor = '#e0e0e0' )
    # change hatch
    #box.set(hatch = '/')
    
plt.show()
```
